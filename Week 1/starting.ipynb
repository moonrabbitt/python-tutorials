### Simple Data Analysis

```python
# Activate the environment
!conda activate coding_3_project

# Let's import some things
import pandas as pd
# your new best friend: https://pandas.pydata.org

# Load the data
df = pd.read_csv('data/ramen-ratings.csv') # careful, read_csv, read_excel is different 
df.head(5)
```

```python
print(f"Shape of the data: {df.shape}") # This is an f string

print(f"Columns of the data: {df.columns}")
```

```python
# Summarise some stuff quickly 
print('Describe the brand column')

print(df.Brand.describe()) # we don't have to use print because it's in a jupyter notebook but if it's a script then print(df.Brand.describe()) - I'll do it for best practice

print('Describe the style column')

print(df.Style.describe())
```

```python
# Let's visualise something

import matplotlib.pyplot as plt

print('Describe the country column')

print(df["Country"].describe())  # this is the same as df.Country.describe()

print('Describe the stars column')

print(df["Stars"].describe())



```

```python
# Let's try plot something
# Spoiler alert it's going to show an error
# if you read the error carefuly you will see it is throwing a type error 
# suggesting that either x or y is not numeric, and to plot a violin plot it has to be.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
sns.violinplot(x="Country", y="Stars", data=df)
plt.xticks(rotation=90)
plt.title("Distribution of Stars by Country")
plt.ylabel("Stars")
plt.xlabel("Country")
plt.show()
```

```python
# Let's see why

print(type(df["Stars"][0])) # Ah! star is a string - Let's fix that

# Convert the stars column to float
df["Stars"] = df["Stars"].astype(float) # we can also use df.Stars.astype(float)

# oh no there are unrated ramen so it didn't work- let's screen that out
```

```python
#Screen out
df = df[df["Stars"] != "Unrated"] 

# Let's try onvert the stars column to float
df["Stars"] = df["Stars"].astype(float) # we can also use df.Stars.astype(float)

print(type(df["Stars"][0])) 
```

```python
# plot spread of countries and stars - violin plot

# https://seaborn.pydata.org
import seaborn as sns
# https://matplotlib.org
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
sns.violinplot(x="Country", y="Stars", data=df)
plt.xticks(rotation=90)
plt.title("Distribution of Stars by Country")
plt.ylabel("Stars")
plt.xlabel("Country")
plt.show()

```
